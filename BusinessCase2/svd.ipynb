# SVD — Singular Value Decomposition

Dimensionality reduction and latent factor analysis for the investment needs dataset.

Typical applications in this context:
- Identify latent client profiles (low-rank approximation of the feature matrix)
- Pre-process features for downstream models
- Collaborative filtering for the recommendation system

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import MinMaxScaler

sys.path.insert(0, str(Path().resolve()))
from utils.preprocessing import FEATURE_NAMES, build_features, load_data

df = load_data()
X = build_features(df)

# Scale before SVD
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
print('Feature matrix shape:', X_scaled.shape)
```

## Explained variance vs number of components

```python
n_components_range = range(1, X_scaled.shape[1] + 1)
explained = []

for n in n_components_range:
    svd = TruncatedSVD(n_components=n, random_state=42)
    svd.fit(X_scaled)
    explained.append(svd.explained_variance_ratio_.sum())

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(list(n_components_range), explained, marker='o')
ax.axhline(0.95, color='red', linestyle='--', label='95% threshold')
ax.set_xlabel('Number of components')
ax.set_ylabel('Cumulative explained variance')
ax.set_title('SVD — Explained Variance')
ax.legend()
plt.tight_layout()
plt.show()
```
